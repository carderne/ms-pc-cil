# Load ISIMIP nc4 data

## Imports

```python
import xarray as xr
```

## Open dataset

```python
path = "/path/to/data.nc4"
```

```python
ds = xr.open_dataset(path)
ds
```

## Map of change from 2021 to 2030

```python
delta = (ds.tas.isel(time=-1) - ds.tas.isel(time=0)).load()
delta.plot(size=6, cmap="RdBu")
```

## Time-series for single location

```python
ts = ds.tas.sel(lon=4, lat=50, method="nearest").load()
ts.plot()
```
